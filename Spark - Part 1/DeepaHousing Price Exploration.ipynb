```python
#import sparkfrom pyspark.sql import SparkSession

from pyspark.sql import SparkSession
```

```python
# Create a SparkSession object

spark = SparkSession.builder \
    .appName("myHousingApp") \
    .getOrCreate()
```

```python
#read in the data set 

housing_data = spark.read.csv('/Users/deepa/Documents/Projects/DataEngineering.Labs.Libraries/Spark - Part 1/AB_NYC_2019.csv',
                              inferSchema=True, header="true")
```

```python
#filter the df to find only listings in Brooklyn using Spark's .filter() method

housing_data.filter(housing_data['neighbourhood_group'] == 'Brooklyn').show(5)
```
